```python
from pathlib import Path

transcriptions_path = Path("~/Google Drive/My Drive/Transcriptions").expanduser().resolve()
obsidian_path = Path("~/Google Drive/My Drive/Obsidian/not_yet_classified").expanduser().resolve()
```

```python
import os
from pydantic import BaseModel, Field
import json

class TopicTranscription(BaseModel):
    title: str
    keywords: list[str]
    summary: str
    content: str
    problem_space: str
    examples: list[str]
    reflection: str
    questions: list[str]

class Metadata(BaseModel):
    source_pages: list[str]
    created_at: str
    class_name: str = Field(alias="class")



topic_transcription, metadata
```

```python
from pprint import pprint
def create_obsidian_file(topic_transcription: TopicTranscription, metadata: Metadata):

    yaml_front_matter = f"""
    ---
    title: "{topic_transcription.title}"
    tags: {topic_transcription.keywords}
    keywords: {topic_transcription.keywords}
    created_at: "{metadata.created_at}"
    source_pages: {metadata.source_pages}
    class: "{metadata.class_name}"
    ---
    """

    body = f"""
    # Problem Space

    {topic_transcription.problem_space}

    # Summary

    {topic_transcription.summary}

    # Content

    {topic_transcription.content}

    # Examples
    """
    # Append each example
    for ex in topic_transcription.examples:
        body += f"\n- {ex}\n"

    body += f"""

    # Reflection

    {topic_transcription.reflection}

    # Questions
    """

    # Append each question
    for q in topic_transcription.questions:
        body += f"\n- {q}\n"

    obsidian_note = yaml_front_matter + body

    new_folder = obsidian_path / metadata.class_name

    new_folder.mkdir(exist_ok=True)

    new_file = new_folder / f"{topic_transcription.title}.md"

    with open(new_file, "w") as f:
        f.write(obsidian_note)
```
